```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas=['Marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget=[2.5, 4, 10, 1, 6]
```

```python
plt.pie(budget, labels=areas, autopct="%0.1f%%")
plt.show()
```

# Scatter plot

```python
x = [23, 76, 34,58, 44, 32, 67]
y = [21, 26, 74,38, 14, 92, 57]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title("Age vs Salary graph")
```

# 2D plot

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [233,456,770,300, 250,450,234,432]
offline_price= [400, 300, 100, 333, 444,222, 199, 900]
chor_bazar = [100, 200, 300, 400, 500, 200, 100, 600]
```

```python
plt.plot(product_id, online_price, marker='D', markersize='10', label='online prices')
plt.plot(product_id, offline_price, color='yellow',markersize='10',label='offline prices')
plt.plot(product_id, chor_bazar, color='green',marker = '+', markersize='10',label='chor prices')
plt.xlabel('Products')
plt.ylabel('Prices')
plt.title("Product price Comparision")
plt.legend()
plt.show()
```

# Bar Graphs

```python
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
price = [65000, 69000, 75000, 55000, 52000, 64000, 81000]
plt.bar(year, price, color='red')
plt.xlabel('Year')
plt.ylabel("Price(In USD)")
plt.title("Year by Year price")
plt.show()
```
